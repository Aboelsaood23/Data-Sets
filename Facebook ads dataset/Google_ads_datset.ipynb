##**Main Dimensions :**
* Campaign Name
* Campaign Status
* Ad Type
* Ad Id
* Start Date

```python
import pandas as pd
import numpy as np
import random
from datetime import datetime,timedelta
data ={
    "Campaign_Name":[],"Campaign_Status":[],"Ad_Type":[],"Ad_ID":[]
}
df =pd.DataFrame(data)
#let's assign values to this dataframe columns
df['Campaign_Name']=['sales','subscrition']
# repeat vales of cmapaign names
index_brand =0
df_repeated =pd.concat([df.loc[[index_brand]]]*156,ignore_index=True).append(df.drop(index_brand),ignore_index=True)
index_leads =156
df_repeated_2 =pd.concat([df_repeated.loc[[index_leads]]]*52,ignore_index=True).append(df_repeated.drop(index_leads),ignore_index=True)
df = df_repeated_2
#Campaign status
Campaign_Status=['Ended','Paused','Eligible','Pending']
np.random.shuffle(Campaign_Status)
df['Campaign_Status'] =df['Campaign_Status'].fillna(pd.Series(np.random.choice(Campaign_Status,size=len(df),p=[1/6,1/3,1/3,1/6])))

Ad_Type=['Search Ads','Shopping Ads','Google Shopping']
df['Ad_Type'] =df['Ad_Type'].fillna(pd.Series(np.random.choice(Ad_Type,size=len(df),p=[1/4,2/4,1/4])))
# add values to ad id to be like this AD_202391,AD_202392 and so on
start_value = "AD_202391"
df['Ad_ID'].fillna(pd.Series(range(len(df))).add(int(start_value[3:])).astype(str).radd("AD_"), inplace=True)
df

# create function to add date column
def generate_random_date():
    start_date = datetime(2023, 1, 9)
    end_date = datetime(2023, 12, 31)
    random_days = random.randint(0, (end_date - start_date).days)
    random_date = start_date + timedelta(days=random_days)
    return random_date

df['Start_Date'] = [generate_random_date() for _ in range(len(df))]
df
```

##**Demographic Targeting Dimensions :**
* Gender
* Age_Target
* Location
* Device

```python
df['Gender'] =np.random.choice(['Male','Female'],size=len(df),p=[0.55,0.45])
df['Age_Target'] =np.random.choice(['18-25','25-35','35-40','45-60','60+'],size =len(df),p=[0.23,0.30,0.20,0.15,0.12])
df['Location'] =np.random.choice(["United States","Brazil","Germany","Australia","Nigeria","France","South Africa","China","Argentina","Canada","Egypt","Japan","New Zealand","Chile"],
                                 size=len(df),p=[0.035110782843004035,0.20773879848777385,0.02800972563880097,0.054192730040288836,0.042251959014462485,0.13120345167648875,0.08903091363761736,
                                                 0.07331957593686136,0.06391963030393041,0.062321639546332165,0.04532482876096884,0.09971462327413146,0.034623133081295644,0.03323820775804382])
df['Device'] =np.random.choice(['Desktop','Mobile','Tablet','Other'],size =len(df),p=[0.23,0.42,0.20,0.15])
df
```

##**Metics :**


*  Amount_spent($)
*  Impressions
*  Clicks
*  Conversions




```python
df['Amount_Spent_$'] =np.random.uniform(100,150, size=len(df))
df['Impressions'] =(df['Amount_Spent_$']*20).astype(int)
df['Clicks'] =(df['Impressions']/8).astype(int)

#ctr = (total_impressions/total_Clicks)*100 "do the calculations on looker "
df['Conversions'] =(df['Clicks']/7).astype(int)

#Calculate value of all conversions
def multiply_conversion(row):
    if row['Ad_Type'] == 'Search Ads':
        return row['Conversions'] * 20
    elif row['Ad_Type'] == 'Shopping Ads':
        return row['Conversions'] * 35
    elif row['Ad_Type'] == 'Google Shopping':
        return row['Conversions'] * 32
    else:
        return None  # Handle other cases if necessary

df['Revenue_$'] = df.apply(multiply_conversion, axis=1)
#Totoal Convesion values =sum(Revenue)  "do this calcculation on looker"
#covnersion rates= (Cinversions/Clicks)*100 "do this calculation on looker"
#cost per conversion =(total_Cost(amount_spent_$)/Conversions) "do this calcaulation on looker"
#RoAS"Rturn on Ad Spend" : total_revenue/total_cost "do this calculation on looker"
#CPC "Cost per Click": totalcost/number of click "do on looker"
#cost per mile "CPM" :total_cost/(numer of impressions/1000) "looker"
df
```
