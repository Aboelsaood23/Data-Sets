###**Main Columns**


*   Campaign Name
*   Ad Set
*   Ad Id



```python
import pandas as pd
import numpy as np
# start by making main columns "campaign,ad_set,ad"
data ={
    'Campaign':[],'Ad_Set':[],'Ad_Id':[]
}
df =pd.DataFrame(data)
#let's assign values to this dataframe columns
df['Campaign']=['brand awarness','generate leads','website traffic']
# repeat vales of cmapaign names
pd.set_option('display.max_rows', None)
index_brand =0
df_repeated =pd.concat([df.loc[[index_brand]]]*300,ignore_index=True).append(df.drop(index_brand),ignore_index=True)
index_leads =300
df_repeated_2 =pd.concat([df_repeated.loc[[index_leads]]]*150,ignore_index=True).append(df_repeated.drop(index_leads),ignore_index=True)
index_traffic =450
df_repeated_3 =pd.concat([df_repeated_2.loc[[index_traffic]]]*150,ignore_index=True).append(df_repeated_2.drop(index_traffic),ignore_index=True)
df = df_repeated_3
df
```

```python
pd.set_option('display.max_rows',10)
ad_set =['May','June','July']
np.random.shuffle(ad_set)
df['Ad_Set'] =df['Ad_Set'].fillna(pd.Series(np.random.choice(ad_set,size=len(df),p=[1/3,1/3,1/3])))
# add values to ad id to be like this AD_468788,AD_468789 and so on
start_value = "AD_468788"
df['Ad_Id'].fillna(pd.Series(range(len(df))).add(int(start_value[3:])).astype(str).radd("AD_"), inplace=True)
df
```

###**Break down columns :**
 ,location,age,start date and end data

*   Gender
*   Location
*   Age Target
*   Start Date
*   End Date





```python
from datetime import timedelta
df['Gender'] =np.random.choice(['Male','Female'],size=len(df),p=[0.65,0.35])
df['Location'] =np.random.choice(["New York City, New York","Los Angeles, California","Chicago,Illinois",
                                           "Houston, Texas","Phoenix, Arizona"],size=len(df), p=[0.23,0.30,0.20,0.15,0.12])
df['Age_Target'] =np.random.choice(['18-25','25-35','35-40','45-60','60+'],size =len(df),p=[0.23,0.30,0.20,0.15,0.12])
# to add dates columns , we will make it on columns
def generate_dates(row):
  start_date =pd.to_datetime(row['Ad_Set']+'1, 2022')+timedelta(days =np.random.randint(0, 30))
  end_date =start_date + timedelta(days =np.random.randint(1,30))
  return pd.Series([start_date,end_date],index=['start_date','end_date'])

df[['Start_date','End_date']]= df.apply(generate_dates,axis =1)
df
```

###**Metrics and dimensions :**
*   Amount Spent($)
*   Reach
*   Impressions
*   Results
*   Cost per result
*   Frequency
*   Post Engagement
*   Post Comments
*   Post Likes
*   post Shares



```python
df['Amount_Spent_$'] =np.random.uniform(10,100, size=len(df))
df['Reach'] =(df['Amount_Spent_$']*30).astype(int) # this for unique users
df['Impressions'] =(df['Reach']*1.4).astype(int)
df['Results'] =np.random.uniform(20,200,size=len(df)).astype(int) #this could be counting for salse
df['Cost_Per_Result'] =(df['Amount_Spent_$']/df['Results'])
df['Frequncy'] =np.random.uniform(1,4,size =len(df)).astype(int)
df['Post_Engagement'] =(df['Impressions']/1.6).astype(int)
df['Post_Comments']=(df['Post_Engagement']/7).astype(int)
df['Post_Shares'] =(df['Post_Engagement']/2-df['Post_Comments']).astype(int)
df['Post_Likes']=(df['Post_Engagement']-df['Post_Comments']-df['Post_Shares'])
df['Follows_Likes']=(df['Post_Engagement']/3).astype(int)
df['Clicks'] =(df['Impressions']/5).astype(int)
df['Cpc_$'] =(df['Amount_Spent_$']/df['Clicks'])
df['Ctr']=(df['Clicks']/df['Impressions'])
```

```python
# convert dataset to csv file
facebook_dataset =df.to_csv("facebook_dataset.csv",index =False)
```
